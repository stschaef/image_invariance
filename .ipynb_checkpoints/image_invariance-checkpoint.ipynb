```python
from matplotlib import pyplot as plt
```

```python
from tensorflow.keras.datasets import mnist
import numpy as np
import scipy
```

```python
(trainX, trainy), (testX, testy) = mnist.load_data()
```

```python
img = trainX[0]
plt.imshow(img, cmap=plt.get_cmap('gray'))
```

```python
img.shape
```

```python
S = np.array([[(1 if i - 1 == j else 0) for j in range(28)]for i in range(28)])
eps = 0.1
S = np.array
```

```python
for i in range(10):
    A = img
    for j in range(i):
        A = np.dot(S, A)
    plt.figure()
    plt.imshow(A, cmap=plt.get_cmap('gray'))
```

```python
shifted = img
for i in range(4):
    shifted = np.dot(S, shifted)
```

```python
t = 0.5
convex = t * img + (1 - t) * shifted 
```

```python
plt.imshow(convex, cmap=plt.get_cmap('gray'))
```

```python
exp_S = scipy.linalg.expm(-0 * S)
plt.imshow(np.dot(exp_S, img), cmap=plt.get_cmap('gray'))
plt.figure()
plt.imshow(img, cmap=plt.get_cmap('gray'))
```
